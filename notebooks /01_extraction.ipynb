# 01 Extraction

Use this notebook to load the raw Swiggy dataset, inspect its structure, and confirm that the source is suitable for the project.

```python
from pathlib import Path
import pandas as pd

current_dir = Path.cwd().resolve()
PROJECT_ROOT = current_dir.parent if current_dir.name.strip() == 'notebooks' else current_dir
```

```python
RAW_PATH = PROJECT_ROOT / 'data/raw/swiggy_data_raw.csv'
df = pd.read_csv(RAW_PATH, parse_dates=['Order Date'])
df.head()
```

```python
print('Rows   :', len(df))
print('Columns:', len(df.columns))
df.info()
```

```python
print('Null counts per column:')
print(df.isnull().sum())
```

```python
print('Duplicate rows:', df.duplicated().sum())
```

```python
print('Unique values per column:')
print(df.nunique())
```

```python
print('Numeric column statistics:')
df.describe()
```

```python
print('Order Date range:')
print('  Min:', df['Order Date'].min())
print('  Max:', df['Order Date'].max())
```
